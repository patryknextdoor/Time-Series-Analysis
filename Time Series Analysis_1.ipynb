```python
import pandas as pd
from datetime import datetime 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn")
#Generating 2 Time Series 
```



##### One way to make the data look more non-stationary is to add a trend to the cumulative sum. You can do this by adding a linear or a polynomial function to the cumulative sum of the random numbers.Another way is to add random noise or fluctuations to the data, for example by adding random values from a normal distribution or by applying a random walk process. You can also consider using different statistical models that can generate non-stationary time series data like ARIMA, GARCH, and so on.

```python
np.random.seed(1)
price1 = np.random.randn(365)
DXY = pd.Series(price1, index=pd.date_range(start="2018", end="2023", periods=365))
DXY = abs(DXY.cumsum())
plt.figure(figsize=(14,7))
DXY.plot()
```

```python
np.random.seed(10)
price2 = np.random.randn(365)

NZD = pd.Series(price2, index=pd.date_range(start="2018", end="2023", periods=365))
NZD = abs(NZD.cumsum())
plt.figure(figsize=(14,7))
NZD.plot()
```

```python
# compute the pearson correlation coefficients
def pearson_r(ts1, ts2):
    #"""Compute Pearson correlation coefficient between the time series."""
    
    # Compute correlation matrix: corr_mat
    corr_mat = np.corrcoef(ts1, ts2)

    # Return entry
    return corr_mat[0,1]

# Compute Pearson correlation coefficient for I. versicolor: r
r = pearson_r(DXY,NZD)

# Print the result
print(f'Pearson correlation between DXY & NZD is {r}')
```

```python
plt.figure(figsize=(14,7))
_ = plt.plot(NZD ,marker='o' ,linestyle="none", color='red')
_ = plt.plot(DXY ,marker='o' ,linestyle="none")
_ = plt.xlabel("YEAR")
_ = plt.ylabel("PRICE")
_
```

```python
# Correlation Matrix

df_corr = df.corr()
df_corr.head()
```

```python
from string import ascii_letters

sns.set_theme(style="white")

# Compute the correlation matrix
df_corr

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(df_corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(df_corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Correlation of Simple Returns
simple_returns = df.pct_change().dropna().corr()
simple_returns.head()
```
